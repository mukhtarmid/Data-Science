```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.decomposition import PCA
pd.set_option('display.max_column', 50)
```

```python
dataInstance = load_breast_cancer()
dataInstance
```

```python
dataInstance.data
```

```python
dataInstance.target
```

```python
dataInstance.DESCR
```

```python
description = dataInstance.DESCR
description.split('\n')
```

```python
data = pd.DataFrame(data=dataInstance.data, columns=dataInstance.feature_names)
data
```

```python
data['target'] = dataInstance.target
```

```python
data.head()
```

```python
data.tail()
```

```python
data.sample(10)
```

```python
X = data.drop(columns='target')
y = data['target']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(X, y, test_size=0.8, random_state=42)
```

```python
decisionTreeClassifier = DecisionTreeClassifier()
```

```python
decisionTreeClassifier.fit(xtrain, ytrain)
```

```python
ypred = decisionTreeClassifier.predict(xtest)
```

```python
accuracy_score(ytest, ypred)
```

```python
print(classification_report(ytest, ypred))
```

```python
pcaInstance = PCA(n_components=30)
```

```python
xtrainPCA = pcaInstance.fit_transform(xtrain)
xtestPCA = pcaInstance.transform(xtest)
```

```python
explainedVarianceRatio = pcaInstance.explained_variance_ratio_
explainedVarianceRatio
```

```python
explainedVariance = pcaInstance.explained_variance_
explainedVariance
```

```python
decisionTreePCAClassifier = DecisionTreeClassifier()
```

```python
pcaNumCompenets = PCA(n_components=5)
```

```python
xtrain5Components = pcaNumCompenets.fit_transform(xtrain)
xtest5Components = pcaNumCompenets.transform(xtest)
```

```python
decisionTreePCAClassifier.fit(xtrain5Components, ytrain)
```

```python
ypredPCA = decisionTreePCAClassifier.predict(xtest5Components)
```

```python
accuracy_score(ytest, ypredPCA)
```

```python
print(classification_report(ytest, ypredPCA))
```
